## Read the .docx file

### Subtask:
Read the content from the "/dataset.docx" file.


**Reasoning**:
Read the content from the "/dataset.docx" file using the Document class from the docx library.



```python
from docx import Document

file_path = "/dataset.docx"

try:
    document = Document(file_path)
    print("Successfully read the document.")
except Exception as e:
    print(f"Error reading the document: {e}")
```

## Process the content

### Subtask:
Depending on the structure of the `.docx` file, process the extracted text to prepare it for conversion to a dataset (e.g., parsing tables, extracting specific information).


**Reasoning**:
Iterate through the tables and rows in the document object, extract text from each cell, and store it in a list of lists.



```python
tables_data = []

for table in document.tables:
    table_data = []
    for row in table.rows:
        row_data = []
        for cell in row.cells:
            row_data.append(cell.text)
        table_data.append(row_data)
    tables_data.append(table_data)

# If there's only one table, flatten the list of lists
if len(tables_data) == 1:
    processed_data = tables_data[0]
else:
    # If there are multiple tables, you might need a different approach to combine them
    # For now, we'll just keep the data from multiple tables separate
    processed_data = tables_data

print("Extracted data from tables.")
```

## Convert to a dataset

### Subtask:
Convert the processed data into a suitable dataset format, like a pandas DataFrame.


**Reasoning**:
Import the pandas library and check the structure of `processed_data` to determine how to create the DataFrame(s).



```python
import pandas as pd

if isinstance(processed_data[0], list):
    # It's a list of lists, check if it's a list of lists of lists
    if isinstance(processed_data[0][0], list):
        # It's a list of lists of lists, multiple tables
        dataframes = [pd.DataFrame(table[1:], columns=table[0]) for table in processed_data]
        print("Created a list of DataFrames from multiple tables.")
    else:
        # It's a list of lists, single table
        df = pd.DataFrame(processed_data[1:], columns=processed_data[0])
        print("Created a single DataFrame from the table.")
else:
    # processed_data is not in an expected format
    print("Processed data is not in a recognized table format.")
    df = None
    dataframes = None
```

## Display the dataset

### Subtask:
Show the first few rows of the created dataset.


**Reasoning**:
Check the type and content of the `dataframes` variable and display the head of the DataFrame(s) accordingly.



```python
if 'dataframes' in locals() and isinstance(dataframes, list) and dataframes:
    for i, dataframe in enumerate(dataframes):
        print(f"Displaying data for Table {i + 1}:")
        display(dataframe.head())
elif 'df' in locals() and df is not None:
    print("Displaying data for the single DataFrame:")
    display(df.head())
else:
    print("No dataset was created or found.")
```
